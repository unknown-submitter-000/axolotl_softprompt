```python
import argparse
import dataclasses
import json
import logging
import math
import pathlib
import random
import sys
import os
import time

import ast
import pandas as pd
import numpy as np
from copy import deepcopy

from tqdm import tqdm
from xopen import xopen

from torch_geometric.data import Data
```

```python
DATA_PATH = "/home/ubuntu/proj/data/CS-TAG"
folder_names = os.listdir(DATA_PATH)
```

```python
for folder_name in folder_names:
    files_names = os.listdir(os.path.join(DATA_PATH, folder_name))
    for file_name in files_names:
        if file_name.endswith('.csv'):
            csv_path = os.path.join(DATA_PATH, folder_name, file_name)
            df = pd.read_csv(csv_path)
            df.neighbour = df.neighbour.apply(lambda x: ast.literal_eval(x))
```

```python
edge_index_i, edge_index_j = [], []
for i in tqdm(df.index):
    node_id = df.loc[i, 'node_id']
    neighbor_ids = df.loc[i, 'neighbour']
    for neighbor_id in neighbor_ids:
        edge_index_i.append(node_id)
        edge_index_j.append(neighbor_id)
edge_index = np.array([edge_index_i, edge_index_j])
```

```python
graph = Data(
    node_text = df.text.values,
    edge_index = edge_index,
)
```
